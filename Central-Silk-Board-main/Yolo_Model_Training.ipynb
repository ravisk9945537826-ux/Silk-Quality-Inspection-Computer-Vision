## YOLOV8m Model

```python
from ultralytics import YOLO
import torch

# Check device
device = 0 if torch.cuda.is_available() else "cpu"

# Load pretrained YOLOv8 model
model = YOLO("yolov8m.pt")

# Train the model
model.train(
    data="Dataset/data.yaml",
    epochs=150,
    imgsz=1024,                 # IMPORTANT for small defects
    batch=4,
    device=device,
    optimizer="AdamW",          # Better for texture-based learning
    lr0=1e-3,
    lrf=1e-2,
    weight_decay=5e-4,
    warmup_epochs=3,
    patience=30,
    workers=8,
    pretrained=True,
    cos_lr=True,                # Smooth LR decay
    amp=True,                   # Mixed precision
    project="silk_defect_yolov8",
    name="yolov8_silk_defects",
    save=True,
    save_period=5,
    verbose=True,
)
```
